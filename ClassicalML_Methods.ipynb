**Classification**

```python
import pandas as pd
df = pd.read_csv('/content/diabetes.csv', sep = ',')
df.head()
df.shape
```

```python
df['Outcome'].value_counts()
```

```python
x = df.drop('Outcome', axis = 1)
x.head()
```

```python
y = df['Outcome']
y
```

```python
x = np.array(x)
y = np.array(y)
x
```

**Normalize**

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x = scalar.fit_transform(x)
x
```

**Train/Test split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

acc_train = accuracy_score (y_train, y_pred_train)
acc_test = accuracy_score (y_test, y_pred_test)
acc_train, acc_test
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix

confusion_matrix(y_test, y_pred_test)
```

```python
p = precision_score(y_test, y_pred_test)
p
```

```python
r = recall_score(y_test, y_pred_test)
r
```

**KNN**: K nearest neighbors


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(x_train, y_train)
```

```python
y_pred_train = knn.predict(x_train)
y_pred_test = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc_train = accuracy_score(y_train, y_pred_train)
acc_test = accuracy_score(y_test, y_pred_test)
acc_train, acc_test
```

```python
from sklearn.metrics import recall_score, precision_score, confusion_matrix

confusion_matrix(y_test, y_pred_test)
```

```python
p = precision_score(y_test, y_pred_test)
p
```

```python
r = recall_score(y_test, y_pred_test)
r
```

Unsupervised ML (K-Means)

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 300, n_features = 2, centers = 3)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], color = 'b')
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3)
model.fit(x)
```

```python
centers = model.cluster_centers_
labels = model.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], c='red', marker = 'x')
plt.show()
```

PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(x_train)

x_new_train = pca.transform(x_train)
x_new_test = pca.transform(x_test)
```

```python
x_train.shape, x_new_train.shape, x_new_test.shape, x_test.shape
```

```python
model = MLPClassifier(hidden_layer_sizes=128, max_iter=200)
model.fit(x_new_train,y_train)
```

```python
y_new_pred_train = model.predict(x_new_train)
y_new_pred_test = model.predict(x_new_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

acc_train = accuracy_score(y_train, y_new_pred_train)
acc_test = accuracy_score(y_test,y_new_pred_test)
acc_train, acc_test
```
